```python
# extract numeric grams from "(30.77g)"
pattern = r"([\d\.]+)\s*(g|gram|grams)"

# create new col
gnc_nutrition["Serving Size (g)"] = (
    gnc_nutrition["Serving Size"]
    .str.extract(pattern, expand=False)[0]
    .astype(float)
)

# remove parentheses in old col
gnc_nutrition["Serving Size"] = (
    gnc_nutrition["Serving Size"]
    .str.replace(pattern, "", regex=True)
    .str.replace(r"\(\s*\)", "", regex=True)
    .str.strip()
)

gnc_nutrition
```
